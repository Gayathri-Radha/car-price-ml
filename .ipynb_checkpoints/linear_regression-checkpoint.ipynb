# Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#2. Read Dataset

df = pd.read_csv("CarPrice_Assignment.csv")
df.head()
```

```python
# 3. Drop Columns That Should Not Be Used

df = df.drop(['car_ID', 'CarName'], axis=1)
df.head()
```

```python
# View columns:
df.columns

# Summary:
df.info()

# Statistics:
df.describe()
```

```python
# 5. One-Hot Encode Categorical Variables
df_encoded = pd.get_dummies(df, drop_first=True)
df_encoded.head()
```

```python
#6. Split Features (X) and Target (y)

X = df_encoded.drop('price', axis=1)
y = df_encoded['price']
```

```python
#7. Scale Features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#8. Train-Test Split (using scaled X)
from sklearn.model_selection import train_test_split

X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
#9. Train Linear Regression Model (scaled)
from sklearn.linear_model import LinearRegression

model_scaled = LinearRegression()
model_scaled.fit(X_train_scaled, y_train)
```

```python
#10. Make Predictions
y_pred_scaled = model_scaled.predict(X_test_scaled)
y_pred_scaled[:10]
```

```python
#11. Evaluate Model
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

r2 = r2_score(y_test, y_pred_scaled)
mae = mean_absolute_error(y_test, y_pred_scaled)
mse = mean_squared_error(y_test, y_pred_scaled)
rmse = np.sqrt(mse)

print("R² Score:", r2)
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
```

```python
# Get Feature Names

feature_names = X.columns
len(feature_names), feature_names[:10]
```

```python
# Get Model Coefficients

coefficients = model_scaled.coef_
coefficients[:10]
```

```python
#Combine Feature Names + Coefficients

feature_importance = pd.DataFrame({
    'Feature': feature_names,
    'Coefficient': coefficients
})
feature_importance.head()
```

```python
#Sort by importance

feature_importance_sorted = feature_importance.sort_values(by='Coefficient', ascending=False)
feature_importance_sorted.head(10)


```

```python
#And the negative ones:

feature_importance_sorted.tail(10)
```

```python
plt.figure(figsize=(12, 14))

# Horizontal bar plot (best for many features)
plt.barh(feature_importance_sorted['Feature'],
         feature_importance_sorted['Coefficient'])

plt.xlabel("Coefficient (Impact on Price)")
plt.ylabel("Feature")
plt.title("Feature Impact on Car Price (Linear Regression)")
plt.axvline(0, color='red', linewidth=1)  # vertical line at 0

plt.show()
```

# Heat map

```python
# -----------------------------------------
# STEP 1: Import necessary libraries
# -----------------------------------------
import seaborn as sns
import matplotlib.pyplot as plt

# -----------------------------------------
# STEP 2: Generate correlation matrix
# (this calculates how strongly each feature relates to others)
# -----------------------------------------
corr = df_encoded.corr()

# -----------------------------------------
# STEP 3: Plot the correlation heatmap
# -----------------------------------------
plt.figure(figsize=(18, 12))  # set the size of the heatmap

sns.heatmap(
    corr,               # correlation matrix
    cmap='coolwarm',    # color theme (red = positive, blue = negative)
    annot=False,        # set True to show numbers inside boxes
    fmt='.2f',          # number format
    linewidths=0.5      # thin lines between cells
)

plt.title("Correlation Heatmap of All Features", fontsize=16)  # title
plt.show()  # display the heatmap
```

# Random Forest Regression

```python
# --------------------------------------------------------------
# STEP 1: Import Random Forest Regressor
# --------------------------------------------------------------
from sklearn.ensemble import RandomForestRegressor
```

```python
# --------------------------------------------------------------
# STEP 2: Train Random Forest Model
# n_estimators = number of trees
# random_state = reproducible results
# --------------------------------------------------------------
rf_model = RandomForestRegressor(
    n_estimators=200,     # number of trees
    random_state=42,
    n_jobs=-1             # use all CPU cores for speed
)

rf_model.fit(X_train_scaled, y_train)
```

```python
# --------------------------------------------------------------
# STEP 3: Predict on Test Data
# --------------------------------------------------------------
rf_pred = rf_model.predict(X_test_scaled)
```

```python
# --------------------------------------------------------------
# STEP 4: Evaluate Model (R2, MAE, RMSE)
# --------------------------------------------------------------
rf_r2 = r2_score(y_test, rf_pred)
rf_mae = mean_absolute_error(y_test, rf_pred)
rf_mse = mean_squared_error(y_test, rf_pred)
rf_rmse = np.sqrt(rf_mse)

print("Random Forest R² Score:", rf_r2)
print("Random Forest MAE:", rf_mae)
print("Random Forest RMSE:", rf_rmse)
```

```python
# --------------------------------------------------------------
# STEP 5: Feature Importance (Tree Based)
# --------------------------------------------------------------
rf_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf_model.feature_importances_
})

rf_importance_sorted = rf_importance.sort_values(
    by='Importance', ascending=False
)

rf_importance_sorted.head(15)
```

```python
# --------------------------------------------------------------
# STEP 6: Plot Feature Importance
# --------------------------------------------------------------
plt.figure(figsize=(12, 10))
plt.barh(rf_importance_sorted['Feature'][:20],
         rf_importance_sorted['Importance'][:20])
plt.gca().invert_yaxis()  # highest important on top
plt.title("Top 20 Feature Importance (Random Forest)")
plt.xlabel("Importance Score")
plt.show()
```

# Decision Tree regression

```python
# --------------------------------------------------------------
# STEP 1: Import Decision Tree Regressor
# --------------------------------------------------------------
from sklearn.tree import DecisionTreeRegressor
```

```python
# --------------------------------------------------------------
# STEP 2: Create the Decision Tree model
# max_depth controls how deep the tree grows (prevents overfitting)
# --------------------------------------------------------------
dt_model = DecisionTreeRegressor(
    max_depth=5,      # limit tree depth (important)
    random_state=42
)

# Train the model
dt_model.fit(X_train_scaled, y_train)
```

```python
# --------------------------------------------------------------
# STEP 3: Predict using the Decision Tree model
# --------------------------------------------------------------
dt_pred = dt_model.predict(X_test_scaled)
```

```python
# --------------------------------------------------------------
# STEP 4: Evaluate Decision Tree
# --------------------------------------------------------------
dt_r2 = r2_score(y_test, dt_pred)
dt_mae = mean_absolute_error(y_test, dt_pred)
dt_rmse = np.sqrt(mean_squared_error(y_test, dt_pred))

print("Decision Tree R²:", dt_r2)
print("Decision Tree MAE:", dt_mae)
print("Decision Tree RMSE:", dt_rmse)
```

# XGBoost Tuning


```python
# --------------------------------------------------------------
# STEP 1: Import GridSearchCV
# GridSearchCV tries different parameter combinations and
# picks the best-performing one.
# --------------------------------------------------------------
from sklearn.model_selection import GridSearchCV
```

```python
# --------------------------------------------------------------
# STEP 2: Define a parameter grid for XGBoost
# Each list contains possible values for that hyperparameter.
# GridSearchCV will try ALL combinations.
# --------------------------------------------------------------
param_grid = {
    'max_depth': [3, 4, 5, 6],       # Tree depth
    'learning_rate': [0.01, 0.05, 0.1],  # How fast model learns
    'n_estimators': [100, 200, 300],     # Number of trees
    'subsample': [0.8, 1.0],             # Sample % of rows
    'colsample_bytree': [0.8, 1.0]       # Sample % of columns
}
```

```python
# --------------------------------------------------------------
# STEP 3: Create basic XGBoost model
# We will tune only selected parameters. Others keep defaults.
# --------------------------------------------------------------
xgb_base = XGBRegressor(
    objective='reg:squarederror',
    random_state=42
)
```

```python
# --------------------------------------------------------------
# STEP 4: Setup GridSearchCV
# cv=3 → 3-fold cross validation
# scoring='r2' → maximize R² score
# n_jobs=-1 → use all CPU cores for speed
# --------------------------------------------------------------
grid_search = GridSearchCV(
    estimator=xgb_base,
    param_grid=param_grid,
    scoring='r2',        # evaluate based on R² score
    cv=3,                # 3-fold cross validation
    n_jobs=-1,           # run in parallel
    verbose=2            # show progress while tuning
)
```

```python
# --------------------------------------------------------------
# STEP 5: Fit GridSearch to training data
# This will take some time because it tests many combinations.
# --------------------------------------------------------------
grid_search.fit(X_train_scaled, y_train)
```

```python
# --------------------------------------------------------------
# STEP 6: Get best parameters and model
# --------------------------------------------------------------
print("Best Parameters:", grid_search.best_params_)
best_xgb = grid_search.best_estimator_
```

```python
# --------------------------------------------------------------
# STEP 7: Evaluate the tuned XGBoost model
# --------------------------------------------------------------
y_pred_best = best_xgb.predict(X_test_scaled)

print("Tuned XGBoost R²:", r2_score(y_test, y_pred_best))
print("Tuned XGBoost MAE:", mean_absolute_error(y_test, y_pred_best))
print("Tuned XGBoost RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_best)))
```

```python
# --------------------------------------------------------------
# Import required libraries
# --------------------------------------------------------------
import shap
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# --------------------------------------------------------------
# Create a TreeExplainer (works best for RandomForest, XGBoost, LightGBM)
# rf_model = your trained Random Forest model
# --------------------------------------------------------------
explainer = shap.TreeExplainer(rf_model)
```

```python
# --------------------------------------------------------------
# SHAP values tell us how much each feature pushes prediction up/down
# Use X_test_scaled because your model was trained on scaled data
# --------------------------------------------------------------
shap_values = explainer.shap_values(X_test_scaled)
```

```python
# --------------------------------------------------------------
# Summary plot shows:
# - Top important features
# - Whether high/low values push price up or down
# - How strong each feature is
# --------------------------------------------------------------

# This initializes Javascript support for interactive visualization
shap.initjs()

# Create summary plot
shap.summary_plot(
    shap_values,           # SHAP values
    X_test_scaled,         # Corresponding test features
    feature_names=X.columns  # Names for the features
)
```
